```python

from flask import Flask, request, jsonify
import numpy as np
from PIL import Image
from fastai.data.all import *
from fastai.vision.all import *
from pathlib import *
from pathlib import Path

app = Flask(__name__)

@app.route('/process_image', methods=['POST'])
def process_image():
    # Verifica se a requisio contm um arquivo de imagem
    if 'file' not in request.files:
        return jsonify({'error': 'Nenhuma imagem encontrada'}), 400

    file = request.files['file']

    # Verifica a extenso do arquivo
    if file.filename == '':
        return jsonify({'error': 'Nome de arquivo invlido'}), 400
    if not file.filename.endswith(('.png', '.jpg', '.jpeg')):
        return jsonify({'error': 'Extenso de arquivo invlida'}), 400


    Image = Image.open(file)

    learn = load_learner('model.pkl')
    Image.show()
    
    nome_classe,indice,probs = learn.predict(Image)

    nome_classe = nome_classe.astype(np.uint8)


    # Retorna o gradiente como JSON
    return jsonify({nome_classe},probs[indice])

if __name__ == '__main__':
    app.run(debug=True)

```
